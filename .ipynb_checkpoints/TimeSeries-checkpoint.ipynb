# Excersies
## Irregular Time Series

In this practical guide we use synthetic time series from `timesynth`[[1]](https://github.com/TimeSynth/TimeSynth).

```python
import timesynth as ts
import numpy as np
import matplotlib.pyplot as plt
import seaborn

import warnings
warnings.filterwarnings('ignore')
```

Let's generate irregular sampling times 

```python
sampler = ts.TimeSampler(stop_time=50)
irregular_times = sampler.sample_irregular_time(num_points=1000, 
                                                keep_percentage=50)
```

then we define the signals we will use later

```python
sinusoid = ts.signals.Sinusoidal(frequency=0.25)
car = ts.signals.CAR(ar_param=0.3, sigma=0.3)
ar_p = ts.signals.AutoRegressive(ar_param=[1.5, -0.75])
pseudo_periodic = ts.signals.PseudoPeriodic(frequency=2, freqSD=0.01, ampSD=0.5)
```

and some white noise $\sim \mathcal{N}(0, \sigma=0.3)$ 

```python
white_noise = ts.noise.GaussianNoise(std=0.3)
```

Finally we sample from each distribution 

```python
signal_list = [sinusoid, car, ar_p, pseudo_periodic]
labels_list = ['sin', 'car', 'ar', 'psp']
```

```python
dataset = dict()
for signal, name in zip(signal_list, labels_list):
    timeseries = ts.TimeSeries(signal, noise_generator=white_noise)
    samples, signals, errors = timeseries.sample(irregular_times)
    x = np.vstack([irregular_times, samples, signals]).T
    dataset[name] = x
```

```python
fig, axes = plt.subplots(len(labels_list), 1, figsize=(10,5), sharex=True)
for i, (key, value) in enumerate(dataset.items()):
    axes[i].errorbar(value[:, 0], value[:, 1], marker='.', label=key)
    axes[i].legend(loc='upper right')

    
axes[-1].set_xlabel('Time')
fig.text(0.06, 0.5, 'Magnitude', va='center', rotation='vertical')
fig.suptitle('Synthetic Time Series', fontsize=15)
```

### Forecasting

We already know about the irregularity in our dataset, so we compare different approaches to address this problem. 

```python
times = dataset['car'][:, 0] # All series were sampled from the same time sampler
dt = np.diff(times)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
seaborn.histplot(dt, ax=axes[0], kde=True)
axes[0].set_title(r'$\Delta t$')

for sig in dataset.values():
    seaborn.kdeplot(sig[:, 1], ax=axes[1], legend=True)

axes[1].legend(list(dataset.keys()))
axes[1].set_title('Magnitudes')
plt.show()
```

### Train Test split

```python
def get_train_test(signal, train_ptg=0.5):
    assert train_ptg >= 0 and train_ptg <= 1, 'train ptg. should be defined in between 0 and 1'
    n_steps = signal.shape[0]
    train_steps = int(n_steps*train_ptg)
    x_train = signal[:train_steps]
    x_test = signal[train_steps:]
    return x_train, x_test
```

## Training Models

For training we use ARIMA model with different `(p, d, q)` combinations. Notice that, when `q=0` and `d=0`, we obtain an AR model and, similarly, we can modify those parameters to achieve MA and ARMA models.  

```python
from statsmodels.tsa.arima_model import ARIMA # univariate
```

First, we search the best conf. for ARIMA model. We support this part using `joblib` to accelerate the searching.

```python
import multiprocessing
from joblib import Parallel, delayed
num_cpu = multiprocessing.cpu_count()
print('{} cores available'.format(num_cpu))
```

In order to estimate the number of lags in the AR model we can visualize the Partial ACF

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

Enders (2014) suggests that with sample size T, the PACF should only be computed up to lags T / 4.

```python
fig, axes = plt.subplots(len(dataset), 2, figsize=(15,10))

for i, (key, value) in enumerate(dataset.items()):
    x_train, x_test = get_train_test(value)

    plot_pacf(x_train[:, 1], 
              lags=len(x_train)/4,
              ax=axes[i][0], 
              title='')
    
    magns = x_train[:, 1]
    axes[i][1].acorr(x_train[:, 1], usevlines=True, 
                     normed=True, maxlags=None, lw=1, label='ACF - {}'.format(key))
    axes[i][1].legend()
    
plt.show()
```

There are no significant contributions from close observations, except for the AR signal. Then we have to try with different lags 

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
def ARIMA_step(data, conf):
    try:
        model = ARIMA(data[:, 1], order=conf)
        model = model.fit(method='css')
        return model
    except Exception as e:
        print(e)
        return None
```

```python
def compute_hypersearch(dataset, fn_step, pdq, scaler_class=MinMaxScaler):
    best_models = dict()

    for key, value in dataset.items():
        # Train and test split
        x_train, x_test = get_train_test(value, train_ptg=0.9)    

        # Standardize signal
        scaler = scaler_class()
        scaler = scaler.fit(x_train)
        x_train = scaler.transform(x_train)
        
        x_test  = scaler.transform(x_test)
        
        try:
            # Prepare distribution
            p = Parallel(n_jobs=num_cpu)
            response = p(delayed(fn_step)(x_train, conf) for conf in pdq)

            aic_list = [(m.aic, m) for m in response if m is not None]
            aic_list = sorted(aic_list, key=lambda tup: tup[0])
            best_arima = aic_list[0][-1]

            # ARIMA
            x_pred = best_arima.forecast(steps=len(x_test))


            best_models[key] = {'model': best_arima,
                                'forecasting': x_pred,
                                'x_test': x_test,
                                'x_train': x_train}
        except Exception as e:
            print(e)
            pass
    return best_models

def plot_results(dataset, results):
    fig, axes = plt.subplots(1, len(dataset.keys()), figsize=(15,5), sharex=True)
    for i, signal_name in enumerate(dataset.keys()):
        model = results[signal_name]['model']
        x_test = results[signal_name]['x_test']
        
        x_train = results[signal_name]['x_train']

        x_pred = results[signal_name]['forecasting'][0]
        x_time = x_test[:, 0]

        axes[i].plot(x_time, x_pred, marker='.', label='Predicted')
        axes[i].plot(x_time, x_test[:, 1], marker='.', label='Original')

        axes[i].set_title('{}'.format(signal_name))

    axes[0].legend(bbox_to_anchor=(2.6, 1.2), ncol=2)
    plt.show()
```

```python
import itertools
p = range(0, 3)
d = range(0, 3)
q = range(0, 3)
pdq = list(itertools.product(p,d,q)) # ALL combinations
```

```python
%%time
ARIMA_results = compute_hypersearch(dataset, ARIMA_step, pdq)
```

```python
plot_results(dataset, ARIMA_results)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

### Defining our model

```python
def create_model():
    inputs = keras.Input(shape=(None, 2), batch_size=1)
    x = layers.LSTM(16, return_sequences=True, name='LSTM')(inputs)
    x = layers.LayerNormalization(name='LayerNorm_0')(x)
    outputs = layers.Dense(1, name='Regressor')(x)
    model = keras.Model(inputs=inputs, outputs=outputs, name="RNN")
    return model
```
