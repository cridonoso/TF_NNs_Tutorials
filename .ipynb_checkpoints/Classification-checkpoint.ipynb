```python
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

## Normalzing Images

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

## Visualizing Dataset

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
fig, axes = plt.subplots(1, 5, figsize=(15,15))
for i in range(5):
    axes[i].imshow(train_images[i])
    axes[i].set_title(class_names[train_labels[i]])
    axes[i].set_xticks([])
    axes[i].set_yticks([])
plt.show()
```

# Machine Learning - Classical Approach

```python
from sklearn.decomposition import NMF
```

```python
img_size = train_images.shape
train_flatten = np.reshape(train_images, [-1, img_size[1]*img_size[2]])
test_flatten = np.reshape(test_images, [-1, img_size[1]*img_size[2]])
```

```python
%%time
n_componentes = 100
nmf = NMF(n_components=n_componentes)
x_train = nmf.fit_transform(train_flatten)
x_test = nmf.transform(test_flatten)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(max_depth=2, random_state=0)
rf_clf.fit(x_train, train_labels)
```

```python
y_pred_rf = rf_clf.predict(x_test)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_recall_rf = f1_score(test_labels, y_pred_rf, average='macro')
```

```python
f1_recall_rf
```

## DeepLearning - Convolutional Neural Network

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow import keras
```

```python
def create_model(img_size, n_classes):
    inputs = Input(img_size)
    x = Conv2D(32, (3, 3), activation='relu')(inputs)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = Flatten()(x)
    x = Dense(64, activation='relu')(x)
    x = Dense(n_classes, activation='softmax')(x)
    
    return keras.Model(inputs=inputs, outputs=x, name="CNN")
```

```python
n_classes = len(class_names)
x_train = tf.expand_dims(train_images, -1)
x_test = tf.expand_dims(test_images, -1)

y_train = tf.one_hot(train_labels, n_classes)
```

```python
model = create_model(x_train.shape[1:], n_classes)
```

```python
keras.utils.plot_model(model, "CNN.png")
```

```python
model.compile(
    loss=keras.losses.CategoricalCrossentropy(),
    optimizer=keras.optimizers.RMSprop(),
    metrics=["accuracy"],
)
```

```python
history = model.fit(x_train, 
                    y_train, 
                    batch_size=64, 
                    epochs=2, 
                    validation_split=0.2)
```

```python
y_probas = model.predict(x_test)
y_pred_cnn = tf.argmax(y_probas, 1) 
```

```python
f1_recall_cnn = f1_score(test_labels, y_pred_cnn, average='macro')
```

```python
from sklearn.utils.multiclass import unique_labels
from sklearn.metrics import confusion_matrix
```

```python
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          ax=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    return ax
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,6))

# Plot normalized confusion matrix
ax0 = plot_confusion_matrix(test_labels, y_pred_rf, classes=np.array(class_names), normalize=True, ax=axes[0],
                      title='RF')

ax1 = plot_confusion_matrix(test_labels, y_pred_cnn, classes=np.array(class_names), normalize=True, ax=axes[1],
                      title='RF')

fig.tight_layout()
```
